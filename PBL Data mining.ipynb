```python
import pandas as pd
import numpy as np

# Fungsi untuk mengisi nilai NaN dengan nilai random dalam rentang maksimal dan minimal per negara
def fill_na_with_random(group):
    for column in group.columns:
        if group[column].dtype in [np.float64, np.int64] and group[column].isnull().any():
            # Filter out inf and -inf values
            valid_values = group[column][np.isfinite(group[column])]
            if not valid_values.empty:
                min_val = valid_values.min()
                max_val = valid_values.max()
                
                if min_val == max_val:
                    # Jika min dan max sama, gunakan sedikit variasi
                    min_val -= 0.1
                    max_val += 0.1
                
                group[column] = group[column].apply(
                    lambda x: np.random.uniform(min_val, max_val) if pd.isna(x) else x
                )
    return group

# Terapkan fungsi ini pada setiap grup negara
df_filled = df_filter.groupby('country').apply(fill_na_with_random).reset_index(drop=True)

# Tampilkan dataframe yang sudah diisi
df_filled

df_filled.to_excel("D:/PENS/Semester_4/Data Mining/PBL_Datamining/clean_data.xlsx", index=False)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Menghapus kolom 'country'
dataset = dataset_df.drop(columns=["country"])

# Menghitung korelasi antara Life Ladder dan variabel lainnya
correlation = dataset.corr()['Life Ladder'].sort_values(ascending=False)

# Visualisasi korelasi menggunakan heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(dataset.corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi antara Variabel')
plt.show()

# Visualisasi korelasi antara Life Ladder dan variabel lainnya
plt.figure(figsize=(23, 6))
correlation.drop('Life Ladder').plot(kind='bar', color='skyblue')
plt.title('Korelasi antara Life Ladder dan Variabel Lainnya')
plt.xlabel('Variabel')
plt.ylabel('Korelasi')
plt.xticks(rotation=45)
plt.show()

```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Pisahkan fitur (X) dan target (y)
X = dataset.drop(columns=['Life Ladder'])
y = dataset['Life Ladder']

# Inisialisasi model Regresi Linier
model = LinearRegression()

# Inisialisasi RFE dengan model dan jumlah fitur yang ingin dipilih
rfe = RFE(model, n_features_to_select=17)

# Fit RFE ke data
rfe.fit(X, y)

# Menampilkan fitur yang terpilih
selected_features = X.columns[rfe.support_]

# Visualisasi fitur yang terpilih
plt.figure(figsize=(10, 6))
sns.barplot(x=rfe.ranking_, y=X.columns)
plt.title('Rangking Fitur dengan Recursive Feature Elimination')
plt.xlabel('Ranking')
plt.ylabel('Fitur')
plt.show()

print("Fitur terpilih:")
print(selected_features)
```

```python
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
import numpy as np

# Pisahkan data pelatihan dan data uji
train = life_ladder_series[:'2010']
test = life_ladder_series['2011':]

# Buat model Exponential Smoothing
model = ExponentialSmoothing(train, trend='add', seasonal=None, seasonal_periods=None)
fit = model.fit()

# Buat prediksi untuk data uji
forecast = fit.forecast(steps=len(test))

# Hitung MAE, MAPE, dan RMSE
mae = mean_absolute_error(test, forecast)
mape = mean_absolute_percentage_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))

# Tampilkan hasil
print(f"MAE: {mae}")
print(f"MAPE: {mape}")
print(f"RMSE: {rmse}")
```

## Moving Average

```python
# Hitung Moving Average (misalnya dengan jendela waktu 3 tahun)
window_size = 3
brunei_df['Moving Average'] = brunei_df['Life Ladder'].rolling(window=window_size).mean()

# Plot hasilnya
plt.figure(figsize=(10, 6))
plt.plot(brunei_df['Life Ladder'], label='Data Aktual')
plt.plot(brunei_df['Moving Average'], label='Moving Average', linestyle='--')
plt.legend(loc='best')
plt.title('Moving Average Peramalan untuk Brunei')
plt.xlabel('Tahun')
plt.ylabel('Life Ladder')
plt.show()
```

```python
# Hitung Moving Average (misalnya dengan jendela waktu 3 tahun)
window_size = 10
brunei_df['Moving Average'] = brunei_df['Life Ladder'].rolling(window=window_size).mean()

# Prediksi 10 tahun ke depan
last_moving_avg = brunei_df['Moving Average'].iloc[-1]
forecast_years = pd.date_range(start=brunei_df.index[-1] + pd.DateOffset(years=1), periods=10, freq='Y')
forecast_values = np.full(10, last_moving_avg)

# Buat DataFrame untuk prediksi
forecast_df = pd.DataFrame({'year': forecast_years, 'forecast': forecast_values})
forecast_df.set_index('year', inplace=True)

# Plot hasil
plt.figure(figsize=(10, 6))
plt.plot(brunei_df['Life Ladder'], label='Data Aktual')
plt.plot(brunei_df['Moving Average'], label='Moving Average', linestyle='--')
plt.plot(forecast_df, label='Peramalan', linestyle='-.', color='red')
plt.legend(loc='best')
plt.title('Moving Average Peramalan untuk Brunei')
plt.xlabel('Tahun')
plt.ylabel('Life Ladder')
plt.show()

# Tampilkan DataFrame hasil peramalan
print(forecast_df)
```

```python
# Hitung MAE, MAPE, dan RMSE
actual_values = brunei_df['Life Ladder'].dropna()
predicted_values = brunei_df['Moving Average'].dropna()

mae = mean_absolute_error(actual_values, predicted_values)
mape = np.mean(np.abs((actual_values - predicted_values) / actual_values)) * 100
rmse = np.sqrt(mean_squared_error(actual_values, predicted_values))

print(f"MAE: {mae}")
print(f"MAPE: {mape}")
print(f"RMSE: {rmse}")
```
