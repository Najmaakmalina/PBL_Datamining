### MODEL SES

```python
# Metode Simple Exponential Smoothing (SES)
def simple_exponential_smoothing(series, alpha):
    """
    Simple Exponential Smoothing (SES)
    
    Parameters:
    - series: time series data (pandas Series)
    - alpha: smoothing parameter (0 < alpha < 1)
    
    Returns:
    - smoothed_series: hasil peramalan menggunakan SES
    """
    smoothed_series = [series[0]]  # Inisialisasi smoothed series dengan nilai pertama dari data asli
    
    for t in range(1, len(series)):
        smoothed_value = alpha * series[t] + (1 - alpha) * smoothed_series[-1]
        smoothed_series.append(smoothed_value)
    
    return pd.Series(smoothed_series, index=series.index)

# Menentukan parameter alpha
alpha = 0.2  # Misalnya, alpha = 0.2

# Melakukan SES
smoothed_series = simple_exponential_smoothing(branch_data['total_amount'], alpha)

# Plot data asli dan hasil peramalan
plt.figure(figsize=(12, 6))
plt.plot(branch_data.index, branch_data['total_amount'], label='Data Asli')
plt.plot(smoothed_series.index, smoothed_series, label='Smoothed Series (SES)')
plt.legend()
plt.title('Simple Exponential Smoothing (SES)')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Data aktual untuk seluruh periode
actual_data = branch_data['total_amount']

# Data yang diprediksi menggunakan moving average
predicted_data = smoothed_series

# Menghitung RMSE
rmse = np.sqrt(mean_squared_error(actual_data, predicted_data))
print("Root Mean Squared Error (RMSE):", rmse)

# Menghitung MAE
mae = mean_absolute_error(actual_data, predicted_data)
print("Mean Absolute Error (MAE):", mae)

# Menghitung MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
mape = mean_absolute_percentage_error(actual_data, predicted_data)
print("Mean Absolute Percentage Error (MAPE):", mape)
```

```python
# Metode Simple Exponential Smoothing (SES)
def simple_exponential_smoothing(series, alpha):
    """
    Simple Exponential Smoothing (SES)
    
    Parameters:
    - series: time series data (pandas Series)
    - alpha: smoothing parameter (0 < alpha < 1)
    
    Returns:
    - smoothed_series: hasil peramalan menggunakan SES
    """
    smoothed_series = [series[0]]  # Inisialisasi smoothed series dengan nilai pertama dari data asli
    
    for t in range(1, len(series)):
        smoothed_value = alpha * series[t] + (1 - alpha) * smoothed_series[-1]
        smoothed_series.append(smoothed_value)
    
    return pd.Series(smoothed_series, index=series.index)

# Menentukan parameter alpha
alpha = 0.2  # Misalnya, alpha = 0.2

# Melakukan SES
smoothed_series = simple_exponential_smoothing(branch_data7['total_amount'], alpha)

# Membuat prediksi 14 hari ke depan
last_value = smoothed_series.iloc[-1]
prediksi_14_hari = [last_value]

for i in range(14):
    next_pred = alpha * prediksi_14_hari[-1] + (1 - alpha) * last_value
    prediksi_14_hari.append(next_pred)

# Konversi prediksi menjadi Series dengan indeks tanggal yang sesuai
prediksi_14_hari = pd.Series(prediksi_14_hari[1:], index=pd.date_range(start=branch_data7.index[-1] + pd.Timedelta(days=1), periods=14, freq='D'))

# Plot data asli, hasil peramalan, dan prediksi 14 hari ke depan
plt.figure(figsize=(12, 6))
plt.plot(branch_data7.index, branch_data7['total_amount'], label='Data Asli')
plt.plot(smoothed_series.index, smoothed_series, label='Smoothed Series (SES)')
plt.plot(prediksi_14_hari.index, prediksi_14_hari, linestyle='--', color='red', label='Prediksi 14 Hari ke Depan')
plt.legend()
plt.title('Simple Exponential Smoothing (SES) - 14 Days Prediction')
plt.show()
```
